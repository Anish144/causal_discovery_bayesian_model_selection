```python
import dill
import numpy as np
import matplotlib.pyplot as plt
from pathlib import Path
```

```python
work_dir = Path("/vol/bitbucket/ad6013/Research/gp-causal")
save_dir = Path("results/lin_gauss_exp")
file_name = "lingauss_results_repeats_100_bayesianlr_optimise_genoffset_offset_prior1.p"
with open(work_dir / save_dir / file_name, "rb") as f:
    results = dill.load(f)
```

```python
ss_list = []
bf_list = []
for ss in results.keys():
    x_to_y_list, y_to_x_list = results[ss]
    bayes_factor = []
    for idx in range(len(x_to_y_list)):
       bayes_factor.append((x_to_y_list[idx] /ss ) - (y_to_x_list[idx] /ss  )) 
    ss_list.append(ss)
    bf_list.append(bayes_factor) 
bf_array = np.array(bf_list)
plt.ylabel(r"$ln p(\mathcal{D}| X \rightarrow Y)/samples - ln p(\mathcal{D}| Y \rightarrow X)/samples$")
plt.xlabel("Number of samples")
plt.grid()
for run_idx in range(100):
    plt.plot(ss_list, bf_array[:, run_idx])
# plt.ylim(-100, 1000)
plt.legend()
plt.show()
```
